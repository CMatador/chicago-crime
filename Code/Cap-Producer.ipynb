```python
# dfrdd = transformed data .toJSON()
df_RDD = spark_df.toJSON()

# dflist = dfrdd.collect()
df_LIST = df_RDD.collect()

# convert to list of JSON dictionaries
df_JSON = [json.loads(row) for row in df_LIST]
for row in df_JSON[:2]:
    print(type(row),'\t', row, '\n')
```

```python
import time, json

output = {'successes':[],'failures':[]}
for row in df_JSON:
    msg = json.dumps(row)
    try:
        producer.produce('chi-crime', msg)
        producer.flush()
        output['successes'].append(msg)
    except Exception as e:
        output['failures'].append(e)
    finally:
#         time.sleep(.5)
        pass

# Print relevant messages ===========================
cnt_fail = len(output['failures'])
if cnt_fail > 0:
    print(f'{cnt_fail} failure(s): {output["failures"][0]}')
else:
    cnt_succ = len(output['successes'])
    print(f'{cnt_succ} success(es)! E.g.:\n')
    for msg in output['successes'][:4]:
        print(type(msg),'\t',msg,'\n')
```
